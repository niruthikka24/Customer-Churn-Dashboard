```python
def initial(df):
  df.drop_duplicates(subset=df.columns.difference(['customer_id']),inplace=True)
```

```python
def fillMedian(df,cols):
  for col in cols:
    df[col] = df[col].fillna(df[col].median())
```

```python
def fillMode(df,cols):
  for col in cols:
    df[col] = df[col].fillna(df[col].mode().iloc[0])
```

```python
def voicemsg(df1):
  df1.loc[(df1['number_vm_messages'].isna()) & (df1['voice_mail_plan'] == 'no'),'number_vm_messages'] = 0.0 #bcz without a voice_mail_plan, can't have voice mail messages
  dfwithvoicepln = df1[df1['voice_mail_plan']=='yes']
  df1.loc[(df1['number_vm_messages'].isna()) & (df1['voice_mail_plan'] == 'yes'),'number_vm_messages'] = dfwithvoicepln['number_vm_messages'].median()
```

```python
def iqr(df1,col):
  Q1 = df1[col].quantile(0.25)
  Q3 = df1[col].quantile(0.75)
  IQR = Q3 - Q1
  whisker_width = 1.5
  lower_whisker = Q1 -(whisker_width*IQR)
  upper_whisker = Q3 + (whisker_width*IQR)
  print("Q1: ", Q1)
  print("Q3: ", Q3)
  print("lower_whisker: ",lower_whisker)
  print("upper_whisker: ",upper_whisker)
```

```python
def chargemin(df,charge,min):
  dfsub = df[[min,charge]]
  dfsub['new'] = dfsub[charge]/dfsub[min]
  rate = dfsub['new'].median()
  print(rate)
  print(dfsub.head())
  statval(dfsub,'new')
  return rate
```

```python
def chargeminreplace(df1,charge,min,thresh1,thresh2):
  rate = chargemin(df1,charge,min)
  df1.loc[(df1[min].isna())|(df1[min]>thresh2)|(df1[min]<thresh1),min] = round(df1.loc[(df1[min].isna())|(df1[min]>thresh2)|(df1[min]<thresh1)][charge]/rate,1)
```

```python
def negtopos(df1,cols):
  for col in cols:
    df1.loc[(df1[col] < 0 ), col] = -1 * df1[col]
```
